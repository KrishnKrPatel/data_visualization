```python
x=np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x),color='green')
plt.xlabel('x',fontsize=15,color='red')
plt.tick_params(axis='y',color='green',labelcolor='blue',labelsize='xx-large')
plt.ylabel('sin(x)',fontsize=15,color='red')
```

```python
plt.tick_params(axis='x',color='green',labelcolor='green',labelsize='xx-large')
```

```python
plt.plot(x,np.sin(x),label='sin-curve')
plt.legend()
```

```python
plt.plot(x,np.sin(x),label='sin-curve')
plt.legend()
plt.title('Playing with Plots',fontsize=15,color='blue')
```

# 2 .Lines and Markers

```python
x=np.linspace(start=0,stop=10,num=50)
```

```python
plt.plot(x,np.sin(x),label='sin-curve')
plt.plot(x,np.cos(x),label='cos-curve')

plt.legend()
plt.title('Playing with Plots')
plt.show()
```

```python
plt.plot(x,np.sin(x),label='sin-curve',color='green')
plt.plot(x,np.cos(x),label='cos-curve',color='m')

plt.legend()
plt.title('Playing with Plots')
plt.show()
```

```python
random_array=np.random.randn(20)
```

```python
plt.plot(random_array,color='m')

plt.title('Playing with Plots')
plt.show()
```

```python
plt.plot(random_array,color='m',linestyle=':')
plt.title('Playing with Plots')
plt.show()
```

```python
plt.plot(random_array,color='green',linestyle='--')
plt.title('Playing with Plots')
plt.show()
```

```python
plt.plot(random_array,color='green',linestyle=':',linewidth=3)
plt.title('Playing with Plots')
plt.show()
```

```python
plt.plot(random_array,color='green',marker='d',markersize=10)
plt.title('Playing with Plots')
plt.show()
```

```python
plt.plot(random_array,color='green',marker='d',markersize=10,linestyle='None')
plt.title('Playing with Plots')
plt.show()
```

```python
plt.scatter(range(0,20),random_array,color='green',marker='d')
plt.title('Playing with Plots')
plt.show()
```
